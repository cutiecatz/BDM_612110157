# Remove Value

```python
df_clean = df #ที่สร้างตัวแปรใหม่มาเพราะว่าถ้าไม่สร้างมันจะลบข้อมูลจริงๆของเรา
df_clean = df_clean.drop('Cabin',1)
df_clean = df_clean.drop('Ticket Number',1)
df_clean.columns
```

```python
df_clean = df_clean.dropna(1,'any') # ถ้า column นี้มีค่าว่าง dropทิ้งทั้ง column
```

```python
df_clean = df_clean.dropna(0,'any') # ถ้าแถวนี้มีค่าว่าง dropทิ้งทั้งแถว
```

```python
df_clean.isnull().any()
```

# Replace Value

```python
df_clean['Sex'].unique()
```

```python
df_clean['Sex'] = df_clean['Sex'].replace('Female','F')
df_clean['Sex'] = df_clean['Sex'].replace('Male','M')
df_clean
```

# Replace string values to unique integers

```python
df_clean['Passenger Class'].unique()
```

```python
p_class, levels = pd.factorize(df_clean['Passenger Class']) # factorize เปลี่ยนข้อมูลจาก string เป็นตัวเลข
```

```python
p_class
```

```python
levels
```

```python
df_clean['Passenger Class'] = p_class
df_clean
```

```python
df_clean.dtypes
```

```python
df_clean['Sex'] = pd.factorize(df_clean['Sex'])[0] # 0 คือ p_class , 1 คือ levels
df_clean['Survived'] = pd.factorize(df_clean['Survived'])[0]
df_clean.dtypes
```

# Set index

```python
df_clean = df_clean.set_index('Name') # index  คล้ายๆกับ Primary key
df_clean
```

# Cross Tab Plot

```python
pd.crosstab(df_clean['Passenger Class'],df_clean['Survived']).plot(kind='bar') # สร้างรูปเปรียบเทียบ
```
